# 五王之战分析 - 冰与火之歌

## 简介

五王之战（War of the Five Kings）是著名严肃奇幻小说《冰与火之歌》中的著名内战。这是一场规模空前、波及七大王国的内乱。顾名思义，前后共有五人在战争中称王：乔佛里、史坦尼斯、蓝礼均声称自己是铁王座的合法继承人。除此之外，罗柏·史塔克被北境众封臣推选为北境之王，巴隆·葛雷乔伊亦再度掀起独立大旗，欲摆脱铁王座的统治，自称为铁群岛之王。



本数据集(battles.csv)包含了五王之战期间的战争，它是所有战斗的大集合。该数据是Kaggle中[Game of Thrones](https://www.kaggle.com/mylesoneill/game-of-thrones)的一部分。

数据中的变量含义解释：
```
name: 战争的名称，字符变量。
year: 战争发生的年份，数值变量。
battle_number: 本数据中的unique id，对应每一场独立的战役，数值变量。
attacker_king: 攻击方的国王，"/"表示了国王的更换。例如："Joffrey/Tommen Baratheon"意味着Tomen Baratheon继承了Joffrey的王位，分类变量。
defender_king: 防守方的国王，分类变量。
attacker_1: 攻击方将领，字符变量。
attacker_2: 攻击方将领，字符变量。
attacker_3: 攻击方将领，字符变量。
attacker_4: 攻击方将领，字符变量。
defender_1: 防守方将领，字符变量。
defender_2: 防守方将领，字符变量。
defender_3: 防守方将领，字符变量。
defender_4: 防守方将领，字符变量。
attacker_outcome: 从攻击方角度来看的战争结果，分别有：win, loss, draw，分类变量。
battle_type: 战争的类别。pitched_battle: 双方军队在一个地点相遇并战斗，这也是最基本的战争类别；ambush: 以隐身或诡计为主要攻击手段的战争；siege: 阵地战；razing: 对未设防位置的攻击。分类变量。
major_death: 是否有重要人物的死亡，二进制变量。
major_capture: 是否有重要人物的被捕，二进制变量。
attacker_size: 攻击方力量的大小，并未对骑兵、步兵等士兵种类有所区分，数值变量。
defender_size: 防守方力量的大小，并未对骑兵、步兵等士兵种类有所区分，数值变量。
attacker_commander: 攻击方的主要指挥官。指挥官的名字中并没有包含头衔，不同的指挥官名字用逗号隔开，字符变量。
defender_commander: 防守方的主要指挥官。指挥官的名字中并没有包含头衔，不同的指挥官名字用逗号隔开，字符变量。
summer: 战争是否发生于夏天，二进制变量。
location: 战争发生的地点，字符变量。
region: 战争发生的地域，包括：Beyond the Wall, The North, The Iron Islands, The Riverlands, The Vale of Arryn, The Westerlands, The Crownlands, The Reach, The Stormlands, Dorne，分类变量。
note: 注释，字符变量。

```

## 数据探索分析

### 1.哪一个国王获得了最多的胜利

````python
# In exploratory data analysis, please make sure of using statistics and visualizations
pd.crosstab(GoTdata.attacker_king,GoTdata.attacker_outcome).style.background_gradient(cmap='summer_r')
````

````python
pd.crosstab([GoTdata.attacker_king,GoTdata.defender_king],GoTdata.attacker_outcome).style.background_gradient(cmap='summer_r')
````

````python
sns.set(font_scale=2.5)#调整字体大小
f,ax=plt.subplots(1,2,figsize=(30,15))
sns.countplot('attacker_king',hue='attacker_outcome',data=GoTdata,ax=ax[0])
ax[0].set_title('Win Vs. loss in attacker_king')
sns.countplot('defender_king',hue='attacker_outcome',data=GoTdata,ax=ax[1])
ax[1].set_title('Win Vs. loss in defender_king')
plt.subplots_adjust(wspace=0.2,hspace=0.5)
plt.show()
````

Joffrey 与Tommen Barathen两代皇帝胜率最高，而且喜欢和Robb Stark打来打去，这两个国家可能仇怨不浅啦；Joffrey/Tommen Barathen 进攻Robb Stark 10次，赢了9次，Robb Stark 进攻Joffrey/Tommen Barathen9次，赢了7次，Joffrey/Tommen Barathen要更盛一筹。

````python
#(2)哪种战争类别最频繁，哪个国王最擅长哪种战争。
sns.countplot("battle_type",data=GoTdata)
````

````python
pd.crosstab(GoTdata.battle_type,GoTdata.attacker_outcome).style.background_gradient(cmap='summer_r')
````

````python
pd.crosstab([GoTdata.attacker_king,GoTdata.battle_type],GoTdata.attacker_outcome).style.background_gradient(cmap='summer_r')
````

pitched_battle是最常见的战争方式，在各个场地中均有较高的获胜率，Joffrey/Tom Baratheon各个场地中都游刃有余，实力强劲；Robb Stark比较擅长ambush战场。

````python
#（3）战争的胜利与双方交战人数的关系
pd.crosstab([GoTdata.attacker_king,GoTdata.defender_king,
             GoTdata.attacker_size,GoTdata.defender_size,GoTdata.attacker_size-GoTdata.defender_size],
            GoTdata.attacker_outcome).style.background_gradient(cmap='summer_r')
````

````python
data = GoTdata.dropna(subset = ["attacker_size", "defender_size"]).copy(deep = True)
data = pd.concat([(GoTdata.attacker_size - GoTdata.defender_size).to_frame(), GoTdata[["attacker_outcome"]]], axis = 1, join = "inner")
data.columns=['attacker_size - defender_size','attacker_outcome']
dwin = data[data['attacker_outcome']=='win']
dloss = data[data['attacker_outcome']=='loss']
f, ax = plt.subplots(1, 2, figsize = (15, 7))
ax[0].set_title("win")
ax[0].set_xlabel('difference in army')
ax[0].set_ylabel('No. of war')
ax[1].set_title("loss")
ax[1].set_xlabel('difference in army')
ax[1].set_ylabel('No. of war')
f.suptitle('how difference in army affect the war outcome')
_ = dwin['attacker_size - defender_size'].plot.barh(ax = ax[0], rot = 0,color = ['steelblue'])
_ = dloss['attacker_size - defender_size'].plot.barh(ax = ax[1],color = ['steelblue'])
plt.show()
````

以少胜多的案例也发生挺多的；Joffrey/Tommen Baratheon与Robb stark是两种完全不同的进攻方式，joffrey/Tommen baratheon喜欢人数碾压，而Robb Stark作为进攻方时，喜欢以少胜多。
